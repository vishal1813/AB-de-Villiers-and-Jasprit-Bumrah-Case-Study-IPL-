## Libraries used

```python
import pandas as pd
import matplotlib as plt
import numpy as np
```

```python
#ipl ball by ball dataset(2008-2020)
df=pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')

df=df.dropna(subset=['is_wicket'])
df=df[df['is_wicket']!='-']
df
```

```python
#finding AB de Villiers 
df[df['batting_team']=='Royal Challengers Bangalore'].batsman.unique()
```

```python
#finding Jasprit Bumrah
df[df['bowling_team']=='Mumbai Indians'].bowler.unique()
```

## AB de Villiers vs JJ Bumrah

```python

dfab=df[(df['batsman']=='AB de Villiers') & (df['bowler']=='JJ Bumrah')]
dfab
```

## Head to Head

```python
runs_scored=sum(dfab.total_runs)
wickets_taken=sum(dfab.is_wicket)
strike_rate=100*sum(dfab.total_runs)/len(dfab)
runs_scored,wickets_taken,strike_rate
```

## Most runs scored by AB de Villiers against individual bowlers

```python
df[df.batsman=='AB de Villiers'].groupby(['bowler'])['total_runs'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

## Most runs scored by individual batsmen against Jasprit Bumrah

```python
df[df['bowler']=='JJ Bumrah'].groupby(['batsman'])['total_runs'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```

## AB de Villiers 

```python
dfa=df[df['batsman']=='AB de Villiers']
dfa
```

## Jasprit Bumrah

```python
dfb=df[df['bowler']=='JJ Bumrah']
dfb
```

## Most balls faced by AB de Villiers agaisnt individual bowlers

```python

faced=pd.DataFrame(dfa['bowler'].value_counts())
faced.head(5)
```

## Most balls bowled by Jasprit Bumrah agaisnt individual batsmen

```python
bowled=pd.DataFrame(dfb['batsman'].value_counts())
bowled.head(5)
```

```python
b1=pd.DataFrame(dfb.groupby('batsman')['ball'].count())
b2=pd.DataFrame(dfb.groupby('batsman')['batsman_runs'].sum())
b3=b1.merge(b2,on='batsman',how='left')
```

## Batsmen with highest strike rate against Bumrah

```python
b3['stike rate']=100*b3.batsman_runs/b3.ball
pd.DataFrame(b3[b3.ball>=30]['stike rate'].sort_values(ascending=False)).head(5)
```

## AB de Villiers's highest strike rate against individual bowlers

```python
a1=pd.DataFrame(dfa.groupby('bowler')['ball'].count())
a2=pd.DataFrame(dfa.groupby('bowler')['batsman_runs'].sum())
a3=a1.merge(a2,on='bowler',how='left')
a3['stike rate']=100*a3.batsman_runs/a3.ball
pd.DataFrame(a3[a3.ball>=30]['stike rate'].sort_values(ascending=False)).head(15)
```

## Batsmen with lowest strike rate against Bumrah

```python
pd.DataFrame(b3[b3.ball>=30]['stike rate'].sort_values(ascending=True)).head(5)
```

## AB de Villiers's lowest strike rate against individual bowlers

```python
pd.DataFrame(a3[a3.ball>=30]['stike rate'].sort_values(ascending=True)).head(5)
```

## Jasprit Bumrah's best bowling strike rate against batsmen

```python
b1=pd.DataFrame(dfb.groupby('batsman')['ball'].count())
b4=pd.DataFrame(dfb.groupby('batsman')['is_wicket'].sum())
b5=b1.merge(b4,on='batsman',how='left')
b5['bowling strike rate']=b5.ball/b5.is_wicket
pd.DataFrame(b5[b5.ball>=30]['bowling strike rate'].sort_values(ascending=True)).head(5)
```

## Bowlers with the best bowling strike rate against AB de Villiers

```python
a1=pd.DataFrame(dfa.groupby('bowler')['ball'].count())
a4=pd.DataFrame(dfa.groupby('bowler')['is_wicket'].sum())
a5=a1.merge(a4,on='bowler',how='left')
a5['bowling strike rate']=a5.ball/a5.is_wicket
pd.DataFrame(a5[a5.ball>=30]['bowling strike rate'].sort_values(ascending=True)).head(5)
```
